```python
import pandas as pd
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
```

```python
dir = r'C:\Users\M.Srimanth\OneDrive\Pictures\PetImages'
classes = ['Cat','Dog']
```

```python
x = []
y = []
for cls in classes:
    path = os.path.join(dir,cls)
    label = classes.index(cls)
    
    for img in os.listdir(path):
        imgpath = os.path.join(path,img)
        try:
            pet_img = cv2.imread(imgpath,0)
            pet_image = cv2.resize(pet_img,(200,200))
        except Exception :
            pass
        x.append(pet_image)
        y.append(label)
```
